```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from bs4.dammit import EncodingDetector
```

```python
driver = webdriver.Chrome(executable_path=r"chromedriver.exe")
driver.get("https://auth.espol.edu.ec/login?service=https%3A%2F%2Fcenacad.espol.edu.ec%2Findex.php%2Fmodule%2FForm%2Faction%2FList")
```

```python
urlComments = "https://cenacad.espol.edu.ec/index.php?module=IntranetReport&action=ComentariosProfesor&id_p=<id_p>&id_e=<id_e>"
```

```python
with open("subjects.txt","r", encoding="utf-8") as subjectsFile:
    lines = subjectsFile.readlines()
```

```python
for line in lines:
    elementos = line.split(",")
    id_p,id_e = elementos[-2],elementos[-1]
    cod,par,anio,sem = elementos[3],elementos[5],elementos[1],elementos[2]
    
    urlCmts = urlComments.replace("<id_p>",id_p).replace("<id_e>",id_e)
    driver.get(urlCmts)
    
    
    html=BeautifulSoup(driver.page_source, 'lxml',from_encoding="utf-8")
    tables = html.findAll("table", {"class": "tbl"})
    heads = html.findAll("td", {"class": "head2"})
    
    if len(tables) == 2:
        table = tables[-1]
        head = heads[0]
        datosProfesor = head.get_text()
        
        tds = table.findAll("td")
        
        idxNombre = datosProfesor.index("Profesor:")
        
        with open( "detalle/"+cod+par+anio+sem+".txt", "a", encoding="utf-8") as detalleFile:
            
            print( cod+par+anio+sem, datosProfesor[idxNombre+len("Profesor:"):].strip() )
            detalleFile.write(datosProfesor[idxNombre+len("Profesor:"):].strip() + "\n")
            
            for td in tds:
                cmtText= td.get_text().strip()

                if "Comentarios Varios" not in cmtText:
                    detalleFile.write(cmtText + "\n")
                    
```
