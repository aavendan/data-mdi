```python
def loadRetweets(candidate):
    rtURL = "https://twitter.com/i/activity/retweeted_popup?id={}"
    
    with open('tweets/ids/{}.txt'.format(candidate),'r') as infile:
        for line in infile:
            
            line = line.strip()
            url = rtURL.format(line.strip())
            
            response = urllib.request.urlopen(url).read()
            jsonDict = json.loads(response)
            
            times = jsonDict['htmlTitle'].split(" ")[1].strip()
            usersHTML = jsonDict['htmlUsers'].replace('\n','').replace(' ',' ')
            
            parsed_html = BeautifulSoup(usersHTML)
            allusers = parsed_html.body.findAll('span', attrs={'class':'username u-dir u-textTruncate'})
            
            users = set([user.text for user in allusers])
                
            results = open('tweets/rtsusers/{}.txt'.format(candidate),'a')
            for user in users:
                results.write( '|'.join([ str(line), str(times), str(len(users)), user])+'\n' )
            results.close()
            
            print(str(line), str(times), str(len(users)))
            
            time.sleep(rd.randint(0,2))
```

```python
for candidate in candidates:
    loadRetweets(candidate)
```

```python
def loadRTUser(candidate, start, end, user):
    
    tries = 0
    
    while True:
        tweetCriteria = got.manager.TweetCriteria().setUsername(user)\
                                               .setSince(start)\
                                               .setUntil(end)\
                                               .setMaxTweets(1000)
        tweets = got.manager.TweetManager.getTweets(tweetCriteria)
        
        tries += 1
        
        if len(tweets) > 1:
            print(user + " " +str(len(tweets)))
       
            for tweet in tweets:

                info = [ 
                    str(tweet.id), 
                    str(user), 
                    str(tweet.date.strftime("%I:%M - %d %b. %Y").lower()), 
                    str( 1 if 'https://twitter.com/{}'.format(candidate) in tweet.text else 0 ), 
                    tweet.text.replace('|','')  ]

                txt = '|'.join(info)+'\n'

                if 'https://twitter.com/{}'.format(candidate) in tweet.text:
                    with open('tweets/rtstext/{}-wh.txt'.format(candidate,user), 'a', encoding="utf-8") as withText:
                        withText.write(txt)
                else:
                    with open('tweets/rtstext/{}-wo.txt'.format(candidate,user), 'a', encoding="utf-8") as withoutText:
                        withoutText.write(txt)
            
            return 
            
        if tries == 10:
            return
```

```python
def loadRTUsers(candidate, start, end):
    
    data = pd.read_csv('tweets/rtsusers/{}.txt'.format(candidate),sep="|", names=['tweetIDCandidate', 'RTNumber', 'RTExtracted', 'User'])
    simpleData = data[['tweetIDCandidate', 'User']]
    uniqueUsers = list(simpleData['User'].unique())
    
    print("Candidate {}".format(candidate))
    print("Users with RTs to crawl: {}".format(len(uniqueUsers)))
    
    func = partial(loadRTUser, candidate, start, end)

    number = 10
    pool = ThreadPool(number)
    time_begin = time.time()
    pool.map(func, uniqueUsers)
    time_end = time.time()
    print("{} thread pool took {} seconds".format(number,time_end - time_begin))
    
```

```python
for candidate in candidates[2:]:
    usuarios = loadRTUsers(candidate, start, end)    
```

```python
def loadLikes(candidate):
    
    likeURL = "https://twitter.com/i/activity/favorited_popup?id={}"
    
    with open('tweets/ids/{}.txt'.format(candidate),'r') as infile:
        for line in infile:
            
            line = line.strip()
            url = likeURL.format(line.strip())
            
            response = urllib.request.urlopen(url).read()
            jsonDict = json.loads(response)
            
            times = jsonDict['htmlTitle'].split(" ")[-2][5:].strip()
            usersHTML = jsonDict['htmlUsers'].replace('\n','').replace(' ',' ')
            
            parsed_html = BeautifulSoup(usersHTML)
            allusers = parsed_html.body.findAll('span', attrs={'class':'username u-dir u-textTruncate'})
            
            users = set([user.text for user in allusers])
                
            results = open('tweets/likeusers/{}.txt'.format(candidate),'a')
            for user in users:
                results.write( '|'.join([ str(line), str(times), str(len(users)), user])+'\n' )
            results.close()
            
            print(str(line), str(times), str(len(users)))
            
            #time.sleep(rd.randint(0,2))
```

```python
for candidate in candidates:
    loadLikes(candidate)
```

```python
def loadAllUsers():
    path = 'tweets/users'

    files = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if '.txt' in file:
                files.append(os.path.join(r, file))

    users = []
    for f in files:
        print(f)
        data = pd.read_csv(f, sep="|", names=['tweetIDCandidate', 'RTNumber', 'RTExtracted', 'User'])
        users += list(data["User"].unique())
    return users
```

```python
def loadDataUsers(user):
    
    user = user.strip().replace('@','')
    url = 'https://twitter.com/{}'.format(user)
    response = urllib.request.urlopen(url).read()
    
    parsed_html = BeautifulSoup(response)
    dataTweets = parsed_html.body.findAll('span', attrs={'class':'ProfileNav-value'})
    dateJoin = parsed_html.body.findAll('span', attrs={'class':'ProfileHeaderCard-joinDateText js-tooltip u-dir'})
    
    lstData = []
    lstData.append(user)
    lstData.append(str(dateJoin[0]['title']).split('-')[0].strip())
    lstData.append(str(dateJoin[0]['title']).split('-')[1].strip()) #['data-original-title'])
    
    for data in dataTweets[:-1]:
        lstData.append(str(int(data.text.replace(u'\xa0', u' ').replace(' M','000000').replace(' mil','000').replace('.','').replace(',','').strip())))
    
    with open('tweets/users/alldata.txt','a', encoding="utf-8") as dataFile:
            dataFile.write('|'.join(lstData)+'\n')
```

```python
with open('tweets/users/allusers.txt','r', encoding="utf-8") as usersFile:
    
    lines = usersFile.readlines()
    
    number = 1
    pool = ThreadPool(number)
    time_begin = time.time()
    pool.map(loadDataUsers, lines)
    time_end = time.time()
    print("{} thread pool took {} seconds".format(number,time_end - time_begin))    
```
